```python
import pandas as pd
from typing import List
from pprint import pprint
import numpy as np
import re
import jieba
from pprint import pprint
```

```python
train_df = pd.read_excel('../data/train_df.xlsx')
test_df = pd.read_excel('../data/test_df.xlsx')

train_df['ckip_ws'] = train_df['ckip_ws'].apply(lambda x: eval(x))
test_df['ckip_ws'] = test_df['ckip_ws'].apply(lambda x: eval(x))

train_df['text'] = train_df['text'].apply(lambda x: str(x))
test_df['text'] = test_df['text'].apply(lambda x: str(x))
```

```python
train_df['jieba_ws'] = train_df['text'].apply(lambda x: jieba.lcut(x))
test_df['jieba_ws'] = test_df['text'].apply(lambda x: jieba.lcut(x))
```

```python
jieba_ws_set = set([word for sentence in train_df['jieba_ws'].tolist() for word in sentence])
jieba_ws_set = jieba_ws_set.union(set([word for sentence in test_df['jieba_ws'].tolist() for word in sentence]))

ckip_ws_set = set([word for sentence in train_df['ckip_ws'].tolist() for word in sentence])
ckip_ws_set = ckip_ws_set.union(set([word for sentence in test_df['ckip_ws'].tolist() for word in sentence]))
```

```python
from wordseg.wordseg import WordSegment
```

```python
sentence = [content.split('\n') for content in train_df['text'].str.replace('\r', '').tolist()]
sentence = ['\n'.join(sublist) for sublist in sentence]
# sentence = [sen for sublist in sentence for sen in sublist if sen != '']
```

```python
ws = WordSegment('\n'.join(sentence), 
                 max_word_len=5, 
                 min_aggregation=5, 
                 min_entropy=0.4)
```

```python
words_freq_info = ws.word_with_freq
print(len(words_freq_info))
```
